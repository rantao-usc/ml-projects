```python
# Load dataset
import pandas as pd
data = pd.read_csv('banking.csv')

# Have a quick look 
data.head()
```

```python
# Let us group "basic.4y", "basic.9y" and "basic.6y" together and call them "basic".
import numpy as np
data['education']=np.where(data['education'] =='basic.9y', 'Basic', data['education'])
data['education']=np.where(data['education'] =='basic.6y', 'Basic', data['education'])
data['education']=np.where(data['education'] =='basic.4y', 'Basic', data['education'])
```

```python
# Delete 'unknown' value

cat_cols = data.select_dtypes(include=object).columns.tolist()

for col in cat_cols:
    # delete each columns that has value 'unknown' 
    index_names = data[ data[col] == 'unknown' ].index 
  
    # drop these row indexes from dataFrame 
    data.drop(index_names, inplace = True) 
```

```python
data = data.drop("default", axis=1)
```

```python
def auc_score_diff(data, ratio):
    
    """
    This function return the difference of percentage of auc score before and after oversampling using SMOTE.
    """
    assert 1 <= ratio <= 6, "Argument 'ratio' out of range"
    
    # downsample the dataset that defaulted
    from sklearn.utils import resample

    df_no_subscribed = data[data['y'] == 0]

    # create label ratio
    n_samples = round(data['y'].copy().value_counts()[1] * ratio)

    df_no_subscribed_downsampled = resample(df_no_subscribed,
                                 replace=False,
                                 n_samples=n_samples,
                                 random_state=1818)
    
    df_subscribed = data[data['y'] == 1]
    data = pd.concat([df_no_subscribed_downsampled, df_subscribed])
    
    # Split X and y
    X = data.drop('y', axis=1).copy()
    y = data['y'].copy()
    
    # One-hot Encoding
    cat_cols = data.select_dtypes(include=object).columns.tolist()
    X_encoded = pd.get_dummies(X, columns=cat_cols)
    X_encoded.head()
    
    # Split training and testing set
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=0)
    
    # Build a preliminary decision tree
    from sklearn.tree import DecisionTreeClassifier
    clf_dt = DecisionTreeClassifier(random_state=1818)
    clf_dt = clf_dt.fit(X_train, y_train)
    
    # Hyperparameter tuning
    from sklearn.model_selection import cross_val_score
    # using cross-validation to find the best value for alpha
    path = clf_dt.cost_complexity_pruning_path(X_train, y_train) # determine values for alpha
    ccp_alphas = path.ccp_alphas # extract different values for alpha
    ccp_alphas = ccp_alphas[:-1] # exclude the maximum value for alpha

    ## create an array to store the results of each fold during cross validation
    alpha_loop_values = []

    ## For each candidate value for alpha, we will run 5-fold cross validation.
    ## Then we will store the mean and standard deviation of the scores (the accuracy) for each call
    ## to cross_val_score in alpha_loop_values...
    for ccp_alpha in ccp_alphas:
        clf_dt = DecisionTreeClassifier(random_state=1818, ccp_alpha=ccp_alpha)
        scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
        alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])
    
    alpha_results = pd.DataFrame(alpha_loop_values,
                            columns=['alpha', 'mean_accuracy', 'std'])

    # get the value for alpha with max mean_accuracy
    ideal_ccp_alpha=alpha_results.iloc[alpha_results['mean_accuracy'].idxmax()]['alpha']  
    
    ## Build and train a new decision tree, only this time use the optimal value for alpha
    clf_dt_pruned = DecisionTreeClassifier(random_state=1818,
                                          ccp_alpha=ideal_ccp_alpha)
    clf_dt_pruned = clf_dt_pruned.fit(X_train, y_train)    
    
    # Calculate First AUC Score
    from sklearn.metrics import roc_auc_score
    
    fir_score = roc_auc_score(y_test, clf_dt_pruned.predict_proba(X_test)[:, 1])
    
    # Oversampling
    from imblearn.over_sampling import SMOTE

    os = SMOTE(random_state=0)
    columns = X_train.columns

    os_data_X,os_data_y=os.fit_resample(X_train, y_train)
    os_data_X = pd.DataFrame(data=os_data_X,columns=columns)
    os_data_y= pd.DataFrame(data=os_data_y,columns=['y'])
    
    X_train, X_test, y_train, y_test = train_test_split(os_data_X, os_data_y, test_size=0.3, random_state=0)
    
    # Build a preliminary decision tree
    from sklearn.tree import DecisionTreeClassifier
    clf_dt_after = DecisionTreeClassifier(random_state=1818)
    clf_dt_after = clf_dt_after.fit(X_train, y_train)
    
    # Hyperparameter tuning
    from sklearn.model_selection import cross_val_score
    # using cross-validation to find the best value for alpha
    path = clf_dt_after.cost_complexity_pruning_path(X_train, y_train) # determine values for alpha
    ccp_alphas = path.ccp_alphas # extract different values for alpha
    ccp_alphas = ccp_alphas[:-1] # exclude the maximum value for alpha

    ## create an array to store the results of each fold during cross validation
    alpha_loop_values_after = []

    ## For each candidate value for alpha, we will run 5-fold cross validation.
    ## Then we will store the mean and standard deviation of the scores (the accuracy) for each call
    ## to cross_val_score in alpha_loop_values...
    for ccp_alpha in ccp_alphas:
        clf_dt = DecisionTreeClassifier(random_state=1818, ccp_alpha=ccp_alpha)
        scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
        alpha_loop_values_after.append([ccp_alpha, np.mean(scores), np.std(scores)])
    
    alpha_results_after = pd.DataFrame(alpha_loop_values_after,
                            columns=['alpha', 'mean_accuracy', 'std'])

    # get the value for alpha with max mean_accuracy
    ideal_ccp_alpha_after=alpha_results_after.iloc[alpha_results_after['mean_accuracy'].idxmax()]['alpha'] 
    
    ## Build and train a new decision tree, only this time use the optimal value for alpha
    clf_dt_pruned_after = DecisionTreeClassifier(random_state=1818,
                                      ccp_alpha=ideal_ccp_alpha_after)
    clf_dt_pruned_after = clf_dt_pruned_after.fit(X_train, y_train)
    
    # Second AUC Score
    sec_score = roc_auc_score(y_test, clf_dt_pruned_after.predict_proba(X_test)[:, 1])
    
    return (sec_score - fir_score) * 100

```

```python
x = np.arange(1, 6, 0.1)
y = [auc_score_diff(data, ratio) for ratio in x]
```

```python
# Split X and y
X = data.drop('y', axis=1).copy()
y = data['y'].copy()
```

```python
# One-hot Encoding
cat_cols = data.select_dtypes(include=object).columns.tolist()
X_encoded = pd.get_dummies(X, columns=cat_cols)
X_encoded.head()
```

# Build A Classification Tree before Oversampling

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=1818)

# create a decision tree and fit it to the training data
from sklearn.tree import DecisionTreeClassifier
clf_dt = DecisionTreeClassifier(random_state=1818)
clf_dt = clf_dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, clf_dt.predict_proba(X_test)[:, 1])
```

```python
from sklearn.model_selection import cross_val_score
# using cross-validation to find the best value for alpha
path = clf_dt.cost_complexity_pruning_path(X_train, y_train) # determine values for alpha
ccp_alphas = path.ccp_alphas # extract different values for alpha
ccp_alphas = ccp_alphas[:-1] # exclude the maximum value for alpha

## create an array to store the results of each fold during cross validation
alpha_loop_values = []

## For each candidate value for alpha, we will run 5-fold cross validation.
## Then we will store the mean and standard deviation of the scores (the accuracy) for each call
## to cross_val_score in alpha_loop_values...
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=1818, ccp_alpha=ccp_alpha)
    scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])

## Now we can draw a graph of the means and standard deviations of the scores
## for each candidate value for alpha         
alpha_results = pd.DataFrame(alpha_loop_values,
                            columns=['alpha', 'mean_accuracy', 'std'])

alpha_results.plot(x='alpha',
                  y='mean_accuracy',
                  yerr='std',
                  marker='o',
                  linestyle='--')
```

```python
# get the value for alpha with max mean_accuracy
ideal_ccp_alpha=alpha_results.iloc[alpha_results['mean_accuracy'].idxmax()]['alpha']
```

```python
## Build and train a new decision tree, only this time use the optimal value for alpha
clf_dt_pruned = DecisionTreeClassifier(random_state=1818,
                                      ccp_alpha=ideal_ccp_alpha)
clf_dt_pruned = clf_dt_pruned.fit(X_train, y_train)
```

```python
from sklearn.metrics import plot_roc_curve
plot_roc_curve(clf_dt_pruned, X_test, y_test) 
```
