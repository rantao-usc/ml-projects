```python
import numpy 
import tensorflow_hub as hub

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

embed = hub.load("https://tfhub.dev/google/Wiki-words-250/2" )
```

```python
a = embed([" king" ])
b = embed([" man" ]) 
c = embed([" woman" ]) 
d = embed([" queen" ])
```

```python
with open("1-1000.txt") as f:
    wordlist = []
    for line in f:
        wordlist.append(line.strip())
```

### Prove the a−b+c vector is closer to the d vector then the majority (for example 99%) of other English word vectors

```python
from scipy.spatial import distance
vec1 = a-b+c
vec2 = d
# distance between 'vector(King) − vector(Man) + vector(Woman)' and 'vector(Queen)'
dst = distance.euclidean(vec1, vec2) 
dst
```

```python
# Compute how many words in this 1000 words list have closer distance from word 'queen' than 'King-Man+Woman'
count = 0
for ele in wordlist:
    embed_ele = embed([ele])
    
    if distance.euclidean(embed_ele, vec2) < dst:
        count += 1
count
```

### Find the top 5 closest words to the word 'dog'

```python
wv_dog = embed(["dog"])
```

```python
# compute the distance between each words of this 1000 words list and dog
dst_between_dog = [distance.euclidean(embed([ele]), wv_dog) for ele in wordlist] 
dst_between_dog
```

```python
# Find the word indices of words that have top 6 smallest distance 
word_indices = sorted(range(len(dst_between_dog)), key=lambda i: dst_between_dog[i])[:6]
word_indices
```

```python
# Find these word indices correspond to which word in the 1000 word list
for ele in word_indices:
    print(wordlist[ele])
```
